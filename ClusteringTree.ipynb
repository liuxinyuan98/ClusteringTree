# Clustering tree algorithm 

## Tools and methods imports

```python
from ClusteringTreeToolkit import ClusteringTreeMethods as ctm
```

## Simulated 13 kinds of traces

```python
filename = r'./Examples/simu_13kinds.txt'
clusteringTree = ctm.clustering_tree_processing(filePath=filename)
clusteringTree.plot_leaf_nodes(show='hist2d')
```

```python
# the CHI in primary step
clusteringTree.plot_CHI_primary()
```

```python
# data flows in primary step
clusteringTree.plot_child_nodes(show='hist2d')
```

```python
# the CHIs in secondary step
clusteringTree.plot_CHIs_secondary()
```

## Simulated 9 kinds of traces with various position and length

```python
# classify the dataset by single step horizontal mapping and K-Means++ clustering
clusteringTree.single_step_horizontal_clustering(show='hist2d')
```

```python
# classify the dataset by single step vertical mapping and K-Means++ clustering
clusteringTree.single_step_vertical_clustering(show='hist2d')
```
